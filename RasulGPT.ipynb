```python
#importing time
import time

def generate_response(message, history):
    yield "â³ Thinking..."
    time.sleep(1.2)

    message = message.lower()

    if "hello" in message or "hi" in message:
        yield "Hello ðŸ‘‹ I'm your Smart AI Assistant. How can I help you today?"
        return

    if "how are you" in message:
        yield "I'm working perfectly âœ… Thanks for asking! How can I help you?"
        return

    if "your name" in message:
        yield "I am Smart AI Assistant â€” built by Rasul ðŸš€"
        return

    if "ai" in message:
        yield "Artificial Intelligence (AI) helps machines learn, think, and make decisions like humans."
        return

    if "gsoc" in message:
        yield "Google Summer of Code (GSoC) is a global program where students contribute to open-source projects."
        return

    if "2+2" in message:
        yield "2 + 2 = 4 âœ…"
        return

    yield f"You asked: '{message}' â€” interesting! Tell me more."
```
